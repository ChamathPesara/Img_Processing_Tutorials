```python
import cv2
img = cv2.imread('dog.jpg')
from cv2 import cvtColor,COLOR_RGB2BGR
img_rgb = cvtColor(img,COLOR_RGB2BGR)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img_rgb)
plt.title("Displaying image using Matplotlib")
plt.show()
```

Observation : Displaying the image using RGB

```python
import cv2
from cv2 import cvtColor,COLOR_BGR2RGB
import matplotlib.pyplot as plt
img = cv2.imread('cat.jpg')
print("Channel order [BGR] : ",img[0,0])
plt.imshow(img)
plt.title('Displaying image using Matplotlib')
plt.show()
```

```python
img_rgb = cvtColor(img,COLOR_BGR2RGB)
print('Channel Order[RGB] : ',img[0,0])
plt.imshow(img_rgb)
plt.title('Displaying image using Matplotlib')
plt.show()
```

```python
import cv2
from cv2 import COLOR_RGB2GRAY
img = cv2.imread('cat.jpg')
print('Before Convertion : ',img[0,0])
img_gray = cv2.cvtColor(img,COLOR_RGB2GRAY)
cv2.imshow("Grayscale image",img_gray)
print("After Convertion : ",img[0,0])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from cv2 import cvtColor
from cv2 import COLOR_RGB2GRAY
import matplotlib.pyplot as plt
img = cv2.imread('cat.jpg')
print('Before Conversion : ',img[0,0])
img_gray = cv2.cvtColor(img,COLOR_RGB2GRAY)
print('After Conversion : ',img[0,0])
plt.imshow(img_gray,cmap='gray',vmin=0,vmax=255)
plt.title("Display image using Matplotlib img_rgb")
plt.show()
```
